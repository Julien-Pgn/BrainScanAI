```python
# Cellule 2 — Vérification GPU
import torch

# Disponibilité CUDA
cuda_available = torch.cuda.is_available()
print("CUDA disponible        :", cuda_available)

if cuda_available:
    print("Nombre de GPU          :", torch.cuda.device_count())
    print("GPU utilisé            :", torch.cuda.get_device_name(0))
    print("Mémoire totale         :", round(torch.cuda.get_device_properties(0).total_memory / 1e9, 2), "GB")
    print("Mémoire disponible     :", round(torch.cuda.mem_get_info(0)[0] / 1e9, 2), "GB")
    print()

    # Test calcul tenseur sur GPU
    x = torch.randn(1000, 1000).cuda()
    y = torch.randn(1000, 1000).cuda()
    z = torch.mm(x, y)
    print("Test multiplication matricielle sur GPU : OK ✓")
    print("Device du tenseur résultat :", z.device)
else:
    print("Aucun GPU détecté — vérifie que le container est lancé avec --gpus all")
```
